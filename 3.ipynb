```python
%matplotlib inline
from skimage.io import imread, imsave, imshow
import os
from skimage import img_as_float
img = imread('data/pretty_girl.jpg')
```

```python
img_f = img_as_float(img)
```

```python
imshow(img_f)
```

```python
r = img_f[:,:,0]
```

```python
imshow(r)
```

```python
g = img_f[:,:,1]
```

```python
imshow(g)
```

```python
b = img_f[:,:,2]
imshow(b)
```

```python
from numpy import dstack
```

```python
img_combined = dstack(tup=(r,g,b))
```

```python
imshow(img_combined)
```

```python
img_av_grey = (r + g + b) / 3
```

```python
imshow(img_av_grey)
```
